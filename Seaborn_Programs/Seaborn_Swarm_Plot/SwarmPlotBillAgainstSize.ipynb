###### Program to draw a swarm plot of total bill against size for a given dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SeaBornSwarmPlot:
    
    def draw_bill_against_size(self):
        """
        function to draw a swarm plot of total 
        bill against size for a given dataset
        """
        tips_dataset = sns.load_dataset('tips')
        sns.swarmplot(x='size', y='total_bill', data=tips_dataset) 
        plt.show() 
        
if __name__ == "__main__" :
    Swarm = SeaBornSwarmPlot()
    Swarm.draw_bill_against_size()
```
