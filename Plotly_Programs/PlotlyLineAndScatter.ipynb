```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np

class Random:
    
    def scatter_plot_for_random_coordinates(self):
        """
        function to draw a line and scatter plot for 
        random 100 x and y coordinates
        """
        N = 100
        x1 = np.random.randn(N) 
        y1 = np.random.randn(N)
        fig = go.Figure()
        #fig = px.line(x=x1, y=y1)
        fig.add_trace(go.Scatter(x=x1,y=y1,mode='markers',name='markers'))
        fig.add_trace(go.Scatter(x=x1,y=y1-5,mode='lines',name='lines'))
        fig.show()
        
if __name__ == "__main__" :
    Ran = Random()
    Ran.scatter_plot_for_random_coordinates()
```
