```python
import plotly.graph_objects as go
import numpy as np

class Random:
    
    def scatter_plot_for_random_coordinates(self):
        """
        function to draw a scatter plot for 
        random 1000 x and y coordinates
        """
        N = 1000
        x1 = np.random.randn(N)
        y1 = np.random.randn(N)

        fig = go.Figure(data=go.Scatter(x = x1, y = y1, mode='markers'))
        fig.show()
        
if __name__ == "__main__" :
    Ran = Random()
    Ran.scatter_plot_for_random_coordinates()
```
