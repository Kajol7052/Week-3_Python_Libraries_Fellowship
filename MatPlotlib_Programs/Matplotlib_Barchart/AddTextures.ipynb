```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

class TexturesAdd:
    
    def add_textures_to_bar(self):
        """
        function to to add textures(black and white) to bars and wedges. 
        Note: Use bottom to stack the womens bars on top of the mens bars
        """
        my_figure = plt.figure() # plt.figure() is to create a figure object
        patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o","X" ]
        ax = my_figure.add_subplot(111)
        for i in range(len(patterns)):
            ax.bar(i, 3, color='white', edgecolor='black', hatch=patterns[i])

        plt.show()
        
        
if __name__ == "__main__":
    Add = TexturesAdd()
    Add.add_textures_to_bar()
```
