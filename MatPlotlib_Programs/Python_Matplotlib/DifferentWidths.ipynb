```python
import matplotlib.pyplot as plt

class DifferentWidths:
    
    def draw_lines_with_different_widths(self):
        """
        function to plot two or more lines 
        with legends, different widths and colors
        """
        x1 =  [1, 2, 3]
        y1 =  [2, 4, 1]
        plt.plot(x1, y1, color = 'hotpink', lw = 2, label = 'line1-width2') # plot line with different widths
        x2 =  [1, 2, 3]
        y2 =  [2.5, 2.5, 2.5]
        plt.plot(x2, y2,  color = 'cadetblue', lw = 1, label = 'line2-width1') # plot line with different widths
        x3 =  [1, 2, 3]
        y3 =  [4, 1, 3]
        plt.plot(x3, y3,  color = 'k', lw = 4, label = 'line3-width4') # plot line with different widths
        x4 =  [1, 2, 3]
        y4 =  [3, 3, 3]
        plt.plot(x4, y4,  color = 'r', lw = 3, label = 'line4-width3') # plot line with different widths
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Draw two or more lines with different widths') # set title
        plt.legend() # show legend on plot
        plt.show() # display figure/graph
        
if __name__ == "__main__":
    Widths = DifferentWidths()
    Widths.draw_lines_with_different_widths()
```
