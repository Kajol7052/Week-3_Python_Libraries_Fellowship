```python
import matplotlib.pyplot as plt

class CurrentAxis:
    
    def draw_lines_with_diplay_current_axis(self):
        """
        function to display the current axis limits
        values and set new axis values
        """
        x1 =  [1, 2, 3]
        y1 =  [2, 4, 1]
        x2 =  [1, 2, 3]
        y2 =  [2.5, 2.5, 2.5]
        plt.plot(x1, y1, x2, y2) # plot line 
        x3 =  [1, 2, 3]
        y3 =  [4, 1, 3]
        plt.plot(x3, y3) # plot line 
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Display current axis limit and set new axis limits') # set title
        print(plt.axis()) # print current axis values
        plt.axis([0, 4, 0, 8]) # limit of x-axis is 0 to 4 and y-axis is 0 to 8
        plt.show() # display figure/graph


        
if __name__ == "__main__":
    Axis = CurrentAxis()
    Axis.draw_lines_with_diplay_current_axis()
```
