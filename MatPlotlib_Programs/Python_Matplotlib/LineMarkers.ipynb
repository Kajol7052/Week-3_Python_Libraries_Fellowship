```python
import matplotlib.pyplot as plt

class MarkersLine:
    
    def draw_lines_with_line_markers(self):
        """
        function to plot two or more lines
        and set the line markers 
        """
        x1 =  [1, 2, 3]
        y1 =  [2, 4, 1]
        plt.plot(x1, y1, lw = 2, marker = 'o', markersize = '20', mec = 'r', mfc = 'k') # plot line with line markers(markersize, mec, mfc)
        x2 =  [1, 2, 3]
        y2 =  [2.5, 2.5, 2.5]
        plt.plot(x2, y2, lw = 2, marker = '*', ms = '10', mec = 'cadetblue', mfc = 'hotpink' ) # plot line with different widths
        x3 =  [1, 2, 3]
        y3 =  [4, 1, 3]
        plt.plot(x3, y3, lw = 2, marker = 'o', ms = '5', mec = 'k', mfc = 'r') # plot line with different widths
        x4 =  [1, 2, 3]
        y4 =  [3, 3, 3]
        plt.plot(x4, y4, lw = 2, marker = '*', ms = '12.5', mec = 'hotpink', mfc = 'cadetblue') # plot line with different widths
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Draw two or more lines with Line Markers') # set title
        plt.show() # display figure/graph
        
if __name__ == "__main__":
    Markers = MarkersLine()
    Markers.draw_lines_with_line_markers()
```
