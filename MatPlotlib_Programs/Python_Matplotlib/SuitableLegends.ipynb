```python
import matplotlib.pyplot as plt

class SuitableLegend:
    
    def draw_lines_with_legend(self):
        """
        function to plot two or more lines on same
        plot with suitable legends of each line
        """
        x1 =  [1, 2, 3]
        y1 =  [2, 4, 1]
        plt.plot(x1, y1, label = 'line 1') # plot line
        x2 =  [1, 2, 3]
        y2 =  [2.5, 2.5, 2.5]
        plt.plot(x2, y2, label = 'line 2') # plot line 
        x3 =  [1, 2, 3]
        y3 =  [4, 1, 3]
        plt.plot(x3, y3, label = 'line 3') # plot line 
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Draw two or more lines with suitable legends') # set title
        plt.legend() # show legend on plot
        plt.show() # display figure/graph

        
if __name__ == "__main__":
    Legend = SuitableLegend()
    Legend.draw_lines_with_legend()
```
