```python
import matplotlib.pyplot as plt
import pandas as pd
  
class CSVFile:
    
    def draw_line_charts(self):
        """
        function to draw line charts of the financial data of 
        Alphabet Inc. between October 3, 2016 to October 7, 2016
        """
        dataframe = pd.read_csv('fdata.csv', sep = ',', parse_dates = True, index_col = 0) # or we can write parse_dates = True
        dataframe.plot()
        plt.show()

if __name__ == "__main__":
    Draw = CSVFile()
    Draw.draw_line_charts()
```
