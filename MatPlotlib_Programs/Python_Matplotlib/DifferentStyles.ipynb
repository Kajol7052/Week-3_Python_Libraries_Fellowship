```python
import matplotlib.pyplot as plt

class DifferentStyles:
    
    def draw_lines_with_styles(self):
        """
        function to plot two or 
        more lines with different styles
        """
        x1 =  [1, 2, 3]
        y1 =  [2, 4, 1]
        plt.plot(x1, y1, label = 'line1-solid', color = 'hotpink', lw = 2, ls = 'solid') # plot line with different styles
        x2 =  [1, 2, 3]
        y2 =  [2.5, 2.5, 2.5]
        plt.plot(x2, y2, label = 'line2-dotted', color = 'cadetblue', lw = 2, ls = 'dotted') # plot line with different styles
        x3 =  [1, 2, 3]
        y3 =  [4, 1, 3]
        plt.plot(x3, y3, label = 'line3-dashed', color = 'k', lw = 2, ls = 'dashed') # plot line with different styles
        x4 =  [1, 2, 3]
        y4 =  [3, 3, 3]
        plt.plot(x4, y4, label = 'line4-dashed/dotted', color = 'r', lw = 2, ls = '-.') # plot line with different styles
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Draw two or more lines with different styles') # set title
        plt.legend() # show legend on plot
        plt.show() # display figure/graph

        
if __name__ == "__main__":
    Style = DifferentStyles()
    Style.draw_lines_with_styles()
```
