```python
import matplotlib.pyplot as plt
import numpy as np
  
class LineDraw:
    
    def basic_line_draw(self):
        """
        function to draw a line with suitable 
        label in the x axis, y axis and a title
        """
        xpoints = np.array([1, 2, 3])
        ypoints =  np.array([3, 2, 1])
        plt.plot(xpoints, ypoints) # plot line 
        plt.xlabel('x-axis') # set x-axis label 
        plt.ylabel('y-axis') # set y-axis label
        plt.title('Draw line') # set title
        plt.show() # display figure/graph

        
if __name__ == "__main__":
    Draw = LineDraw()
    Draw.basic_line_draw()
```
