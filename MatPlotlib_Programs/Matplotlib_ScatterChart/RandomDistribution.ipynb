```python
import matplotlib.pyplot as plt
import numpy as np

class ScatterPlot:
    
    def random_distribution_scatter_plot(self):
        """
        function to draw a scatter graph taking a random 
        distribution in X and Y and plotted against each other
        """
        x = np.random.randn(100) # randn() func. return a random matrix with data from the "standard normal" distribution
        y = np.random.randn(100)
        plt.scatter(x,y, color='c', alpha=0.3)
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        plt.title("Scatter Graph random distribution")
        plt.show()
        

        
        
if __name__ == "__main__":
    Scatter = ScatterPlot()
    Scatter.random_distribution_scatter_plot()
```
