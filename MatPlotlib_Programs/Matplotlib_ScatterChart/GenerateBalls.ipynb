```python
import matplotlib.pyplot as plt
import numpy as np

class GenerateBalls:
    
    def generate_balls_of_different_sizes(self):
        """
        function to draw a scatter plot using random distributions
        to generate balls of different sizes
        """
        x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
        y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
        sizes = np.array([20,50,100,200,500,100,60,90,10,300,600,400,75])
        plt.scatter(x,y, color='olive', s=sizes)
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        plt.title("Generate Balls of Different Sizes")
        plt.show()
        

        
        
if __name__ == "__main__":
    Generate = GenerateBalls()
    Generate.generate_balls_of_different_sizes()
```
