```python
import matplotlib.pyplot as plt
import numpy as np

class EmptyCircle:
    
    def empty_circles_on_scatter_plot(self):
        """
        function to draw a scatter plot with empty
        circles taking a random distribution in X and Y 
        and plotted against each other
        """
        x = np.random.randn(100) # randn() func. return a random matrix with data from the "standard normal" distribution
        y = np.random.randn(100)
        plt.scatter(x,y, color='c', facecolors='None', s=20)
        plt.xlabel("x-axis")
        plt.ylabel("y-axis")
        plt.title("Empty Circle random distribution")
        plt.show()
        

        
        
if __name__ == "__main__":
    Empty = EmptyCircle()
    Empty.empty_circles_on_scatter_plot()
```
