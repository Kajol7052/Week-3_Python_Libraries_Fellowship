```python
import matplotlib.pyplot as plt

class MultipleWedgesInPie:
    
    def make_multiple_wedges_in_pie(self):
        """
        function to create a pie chart with a title 
        of the popularity of programming Languages.
        Make multiple wedges of the pie.
        """
        programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
        popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        my_colors = ['cadetblue','hotpink', 'peru', 'salmon', 'olive', 'yellow']
        my_explode = [0.1, 0, 0, 0, 0,0]
        plt.pie(popularity, labels=programming_languages, colors=my_colors, explode=my_explode, startangle=180, shadow=True) # pie() is used to plot pie chart
        plt.title("Popularity of Programming Languages", bbox={'facecolor':'1.0'}) # bbox is bounding box
        plt.legend()
        plt.show() 

        
        
if __name__ == "__main__":
    Multiple = MultipleWedgesInPie()
    Multiple.make_multiple_wedges_in_pie()
```
