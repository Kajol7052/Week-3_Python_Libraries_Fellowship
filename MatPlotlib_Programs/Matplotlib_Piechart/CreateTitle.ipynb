```python
import matplotlib.pyplot as plt

class CreateTitleOfPieChart:
    
    def create_title_of_pie_chart(self):
        """
        function to create a pie chart with a title of
        the popularity of programming Languages
        """
        programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
        popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        my_colors = ['cadetblue','hotpink', 'peru', 'salmon', 'olive']
        plt.pie(popularity, labels=programming_languages, colors=my_colors) # pie() is used to plot pie chart
        plt.title("Popularity of Programming Languages", bbox={'facecolor':'1.0'}) # bbox is bounding box
        plt.show() 

        
        
if __name__ == "__main__":
    TitlePie = CreateTitleOfPieChart()
    TitlePie.create_title_of_pie_chart()
```
