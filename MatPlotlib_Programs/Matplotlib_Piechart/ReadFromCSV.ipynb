```python
import matplotlib.pyplot as plt
import pandas as pd

class ReadFromCsv:
    
    def read_data_from_csv_file(self):
        """
        function to create a pie chart of gold medal achievements
        of five most successful countries in 2016 Summer Olympics. 
        Read the data from a csv file.
        """
        dataframe = pd.read_csv('medal.csv')
        country = dataframe['country']
        medal = dataframe['gold_medal']
        my_colors = ['cadetblue','hotpink', 'peru', 'salmon', 'olive']
        my_explode = [0.1, 0, 0, 0, 0]
        plt.pie(medal, labels=country, colors=my_colors, explode=my_explode, startangle=180, shadow=True, autopct='%1.2f%%') # pie() is used to plot pie chart
        plt.title("Gold medal achievements of successful countries", bbox={'facecolor':'1.0'}) # bbox is bounding box and autopct enables you to display the percent value using Python string formatting
        plt.legend(title='Countries: ')
        plt.show() 
        

        
        
if __name__ == "__main__":
    Read = ReadFromCsv()
    Read.read_data_from_csv_file()
```
