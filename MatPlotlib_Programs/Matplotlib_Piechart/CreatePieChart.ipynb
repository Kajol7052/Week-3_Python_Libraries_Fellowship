```python
import matplotlib.pyplot as plt

class CreatePieChart:
    
    def create_pie_chart(self):
        """
        function to create a pie chart of the 
        popularity of programming Languages
        """
        programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
        popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        my_colors = ['cadetblue','hotpink', 'peru', 'salmon', 'olive', 'red']
        plt.pie(popularity, labels=programming_languages, colors=my_colors) # pie() is used to plot pie chart
        plt.show() 

        
        
if __name__ == "__main__":
    Pie = CreatePieChart()
    Pie.create_pie_chart()
```
